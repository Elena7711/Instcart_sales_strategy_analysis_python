### Changes

```python
#renaming columns
df_customers.rename(columns = {'First Name' : 'first_name'},
                inplace = True
               )
df_customers.rename(columns = {'Surnam' : 'last_name'},
                inplace = True
               )
df_customers.rename(columns = {'Gender' : 'gender'},
                inplace = True
               )
df_customers.rename(columns = {'Age' : 'age'},
                inplace = True
               )
df_customers.rename(columns = {'STATE' : 'state'},
                inplace = True
               )
```

#### Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
#check for a mixed data
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

There is a mixed data type in the column 'first_name' 

```python
#find NaN value
print(df_customers['first_name'].isna().sum())
```

```python
#locate Nan 
df_customers.loc[df_customers['first_name'].isna()]
```

```python
# NaN value with 'Unknown'
df_customers['first_name'].fillna('Unknown', inplace=True)
```

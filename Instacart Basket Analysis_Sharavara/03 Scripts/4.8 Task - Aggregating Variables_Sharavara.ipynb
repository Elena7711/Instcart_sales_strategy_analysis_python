##  Create loyalty flag for existing customers.

```python
# Check the data grouping for accuracy
pd.options.display.max_rows = None
```

```python
df.head(100)
```

```python
# Create flags to assign "loyalty" labels to user ID based on max order value
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check frequencies of loyalty flags for accuracy
df['loyalty_flag'].value_counts(dropna = False)
```

## Find the mean of product prices by loyalty group.

```python
# Group by "prices" 
df.groupby('prices')
```

```python
# Split the data into groups by "loyalty_flag" and then calculate the mean for each "prices" column
df.groupby('loyalty_flag').agg({'prices': ['mean']})
```

## Create spending flag for each user based on average price of products across orders

```python
# Group data by "user_id" column, generate mean product price for each user, create "mean_price" for aggregation results
df['mean_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check grouping
df.head(100)
```

```python
# Create flags to assign "spending" labels to user ID based on mean product price
df.loc[df['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df.loc[df['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequencies of spending flags for accuracy
df['spending_flag'].value_counts(dropna = False)
```

##  Create order frequency flag that indicates users' ordering behavior

```python
# Group data by "user_id" column, generate median product of "days since prior order, create "median_days" for aggregation results
df['median_days'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check grouping
df.head(100)
```

```python
# Create flags to assign "freqency" labels to user ID based on median days since prior order
df.loc[df['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df.loc[(df['median_days'] > 10) & (df['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df.loc[df['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequencies of spending flags for accuracy
df['frequency_flag'].value_counts(dropna=False)
```

```python
# Examining the null values
print(df[df.frequency_flag.isnull()])
```

```python
# Changing the null values in "frequency_flag" column to "Non-frequent customer"
df['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
df['frequency_flag'].value_counts(dropna=False)
```
